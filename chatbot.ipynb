```python
df = pd.read_csv("FAQ_chatbot.csv")

# split single column into 3
df = df[df.columns[0]].str.split(",", expand=True)
df.columns = ["question", "answer", "category"]

print(df.head())
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["question"])
```

```python
def chatbot(user_input):
    user_input = user_input.lower()
    
    # Convert input to vector
    user_vec = vectorizer.transform([user_input])
    
    # Calculate similarity
    similarity = cosine_similarity(user_vec, X)
    
    # Get best match
    index = similarity.argmax()
    score = similarity[0][index]
    
    # If no good match
    if score < 0.3:
        return "Sorry, I don't understand. Please ask another question."
    
    return df["answer"][index]
```

```python
for _ in range(20):   # limited loop (no infinite running issue)
    query = input("You: ")
    
    if query.lower() == "exit":
        print("Bot: Goodbye!")
        break
    
    print("Bot:", chatbot(query))
```
